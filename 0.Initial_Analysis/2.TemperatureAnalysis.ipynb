```python
# Convert the 'timestamp' column from epoch time (milliseconds) to datetime objects
# The unit='ms' argument specifies that the timestamps are in milliseconds.
df['IST'] = pd.to_datetime(df['timestamp'], unit='ms')

# Localize the timezone to UTC first, and then convert it to 'Asia/Kolkata' (IST).
df['IST'] = df['IST'].dt.tz_localize('UTC').dt.tz_convert('Asia/Kolkata')

# To save the new DataFrame with the 'IST' column, you can use the following line:
df.to_csv('TemperatureAnalysis_11Aug25_13Aug25_IST.csv', index=False)
```

```python
SELECT
  DATE_TRUNC('DAY', CAST(record_time AS TIMESTAMP)) AS record_time,
  AVG(battery_temperature) AS "AVG(battery_temperature)",
  AVG(motor_temperature) AS "AVG(motor_temperature)"
FROM (
  WITH vehicle_running_data AS (
    SELECT
      id AS bus_id,
      timestamp,
      dt,
      batterycoolanttemperature AS coolant_temp,
      motor_temperature,
      pack1_cellmax_temperature AS avg_pack_temp
    FROM facts_prod.can_parsed_output_100
    WHERE
      dt >= CURRENT_DATE - INTERVAL '30' DAY
      AND ignitionstatus = 1
      AND (
        vehicle_speed_vcu > 0 OR motor_rpm > 0
      )
      AND gun_connection_status = 0
  ), temperature_data AS (
    /* Create unified temperature fields */
    SELECT
      bus_id,
      timestamp,
      COALESCE(coolant_temp, avg_pack_temp) AS battery_temperature, /* Use coolant temperature as primary, pack temp as backup */
      motor_temperature
    FROM vehicle_running_data
    WHERE
      NOT COALESCE(coolant_temp, avg_pack_temp) IS NULL
      AND NOT motor_temperature IS NULL
  )
  SELECT
    bus_id,
    timestamp AS record_time,
    battery_temperature,
    motor_temperature
  FROM temperature_data
  ORDER BY
    bus_id,
    record_time
) AS virtual_table
GROUP BY
  DATE_TRUNC('DAY', CAST(record_time AS TIMESTAMP))
ORDER BY
  "AVG(battery_temperature)" DESC
LIMIT 1000

# fleet_temprature 


# NOTE - later added vehcle ID here - https://naarni.slack.com/archives/D09A1LQBTGQ/p1755632283126879?thread_ts=1755600533.477159&cid=D09A1LQBTGQ
```
