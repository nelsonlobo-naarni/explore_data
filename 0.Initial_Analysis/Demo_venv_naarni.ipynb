```python
# check_latest_gps.py
import trino
import pandas as pd

def run_query(query):
    try:
        conn = trino.dbapi.connect(
            host="trino",
            port=8080,
            user="admin",
            catalog="adhoc",
            schema="facts_dev"
        )
        cursor = conn.cursor()
        cursor.execute(query)
        columns = [desc[0] for desc in cursor.description]
        data = cursor.fetchall()
        df = pd.DataFrame(data, columns=columns)
        cursor.close()
        conn.close()
        return df
    except Exception as e:
        print(f"Error: {e}")
        return None

# Check if the view exists (using correct Trino syntax)
print("Checking if view exists:")
query1 = "SHOW TABLES FROM facts_dev LIKE 'latest_gps'"
df1 = run_query(query1)
if df1 is not None:
    print(df1)

# Query the view to see the data
print("\nQuerying the view:")
query2 = "SELECT * FROM facts_dev.latest_gps LIMIT 10"
df2 = run_query(query2)
if df2 is not None:
    display(df2)
```
